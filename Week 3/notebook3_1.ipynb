## Random Walk

```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import datetime
import random
import re

# Seed for reproducibility
random.seed(datetime.datetime.now().timestamp())

def getLinks(articleUrl):
    base = "https://en.wikipedia.org"
    # Always use HTTPS and set headers
    req = Request(base + articleUrl, headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req)
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id': 'bodyContent'}).find_all(
        'a', href=re.compile(r"^(/wiki/)((?!:).)*$"))

# Start with Kevin Bacon page
links = getLinks('/wiki/Kevin_Bacon')

while len(links) > 0:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print("https://en.wikipedia.org" + newArticle)
    links = getLinks(newArticle)
```

## Recursively crawling an entire site

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import re, time

pages = set()
BASE = "https://en.wikipedia.org"
HDRS = {'User-Agent': 'Mozilla/5.0'}

def getLinks(pageUrl):
    try:
        req = Request(BASE + pageUrl, headers=HDRS)
        html = urlopen(req)
        bs = BeautifulSoup(html, 'html.parser')
    except Exception as e:
        print("skip:", pageUrl, e)
        return

    for a in bs.find_all('a', href=re.compile(r"^/wiki/")):
        href = a.get('href')
        if not href:
            continue
        # skip special pages like File:, Help:, etc.
        if ":" in href.split("/wiki/")[-1]:
            continue
        if href not in pages:
            pages.add(href)
            print(BASE + href)
            time.sleep(0.2)  # be nice
            getLinks(href)

# start from home (or e.g. '/wiki/Kevin_Bacon')
getLinks('')
```

## Collecting Data Across an Entire Site

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import re, time, collections

BASE = "https://en.wikipedia.org"
HDRS = {'User-Agent': 'Mozilla/5.0'}

def fetch(url):
    req = Request(url, headers=HDRS)
    with urlopen(req) as resp:
        return resp.read()

def get_first_paragraph(bs):
    bc = bs.find('div', id='bodyContent')
    if not bc:
        return None
    # prefer the lead paragraph in the main content area
    p = bc.select_one('div.mw-parser-output > p')
    return p

def edit_href(bs):
    # edit tab when editable; viewsource when protected; may be missing
    li = bs.find(id='ca-edit') or bs.find(id='ca-viewsource')
    if li and li.a and li.a.has_attr('href'):
        return BASE + li.a['href']
    return None

def crawl(start='/wiki/General-purpose_programming_language', max_pages=50):
    seen = set()
    dq = collections.deque([start])
    while dq and len(seen) < max_pages:
        page = dq.popleft()
        if page in seen:
            continue
        seen.add(page)

        try:
            html = fetch(BASE + page)
            bs = BeautifulSoup(html, 'html.parser')
        except Exception as e:
            print("skip:", page, e)
            continue

        # Title
        title = bs.h1.get_text(strip=True) if bs.h1 else "(no title)"
        print(title)

        # First paragraph (if any)
        p = get_first_paragraph(bs)
        if p:
            print(p.get_text(strip=True))

        # Edit/View source link (if any)
        eh = edit_href(bs)
        print(eh if eh else "(no edit/viewsource link)")

        # Enqueue internal article links (skip special namespaces with ':')
        for a in bs.find_all('a', href=re.compile(r"^/wiki/")):
            href = a.get('href')
            if not href or ":" in href.split("/wiki/")[-1]:
                continue
            if href not in seen:
                dq.append(href)

        print("-" * 20)
        time.sleep(0.2)  # be polite

crawl()
```

## Crawling across the Internet

```python
from urllib.request import Request, urlopen
from urllib.parse import urlparse, urljoin
from bs4 import BeautifulSoup
import re
import time
import random
import socket

USER_AGENT = "Mozilla/5.0"
TIMEOUT = 15
SLEEP = 0.3
MAX_HOPS = 50

def fetch_bs(url):
    """Fetch a URL with headers and return BeautifulSoup, or None on failure."""
    try:
        req = Request(url, headers={"User-Agent": USER_AGENT})
        with urlopen(req, timeout=TIMEOUT) as resp:
            # Basic content-type guard
            ctype = resp.headers.get("Content-Type", "")
            if "text/html" not in ctype:
                return None
            html = resp.read()
        return BeautifulSoup(html, "html.parser")
    except (Exception, socket.timeout) as e:
        print("skip:", url, "-", e)
        return None

def get_internal_links(bs, base_url):
    """Return a set of absolute internal links for base_url."""
    if not bs:
        return set()
    origin = urlparse(base_url)
    origin_netloc = origin.netloc
    origin_scheme = origin.scheme
    internal = set()
    for a in bs.find_all("a", href=True):
        href = a["href"].strip()
        # Build absolute URL
        abs_url = urljoin(base_url, href)
        p = urlparse(abs_url)
        if p.scheme in ("http", "https") and p.netloc == origin_netloc:
            # Normalize (remove fragments)
            abs_url = abs_url.split("#", 1)[0]
            internal.add(abs_url)
    return internal

def get_external_links(bs, base_url):
    """Return a set of absolute external links for base_url."""
    if not bs:
        return set()
    origin_netloc = urlparse(base_url).netloc
    external = set()
    for a in bs.find_all("a", href=True):
        href = a["href"].strip()
        abs_url = urljoin(base_url, href)
        p = urlparse(abs_url)
        if p.scheme in ("http", "https") and p.netloc and p.netloc != origin_netloc:
            abs_url = abs_url.split("#", 1)[0]
            external.add(abs_url)
    return external

def get_random_external_link(starting_page):
    """Get a random external link reachable from starting_page (walks internally if needed)."""
    bs = fetch_bs(starting_page)
    if not bs:
        return None
    externals = get_external_links(bs, starting_page)
    if externals:
        return random.choice(list(externals))
    # No externals here → try random internal
    internals = get_internal_links(bs, starting_page)
    if not internals:
        return None
    next_internal = random.choice(list(internals))
    return get_random_external_link(next_internal)

def follow_external_only(starting_site, max_hops=MAX_HOPS):
    """Follow random external links up to max_hops."""
    current = starting_site
    visited = set([current])
    for hop in range(1, max_hops + 1):
        nxt = get_random_external_link(current)
        if not nxt:
            print("No further external links found from:", current)
            break
        if nxt in visited:
            print(f"[{hop}] (cycle) {nxt}")
        else:
            print(f"[{hop}] {nxt}")
            visited.add(nxt)
        current = nxt
        time.sleep(SLEEP)

follow_external_only("https://www.oreilly.com/")
```
